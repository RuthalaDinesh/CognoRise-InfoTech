```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv("Unemployment in India.csv")
df
```

```python
df.shape

```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Date"]=pd.to_datetime(df[" Date"],dayfirst=True)
df
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
num_df=df.select_dtypes(include=["int64","float64"])
num_df
```

```python
sns.heatmap(num_df.corr(),annot=True,cmap="YlGnBu")
```

```python
df["Year"]=df["Date"].dt.year
df["Month"]=df["Date"].dt.month 
df.head()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(14,8))
sns.barplot(x="Region",y=" Estimated Employed",hue="Region",data=df,palette="YlGnBu")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
df.groupby("Area")[" Estimated Unemployment Rate (%)"].mean().plot(kind="pie",autopct="%.2f")
plt.title("Area wise Unemployment Rate")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df.groupby("Area")[" Estimated Employed"].mean().plot(kind="pie",autopct="%.2f")
plt.title("Area wise Employment Rate")
plt.show()
```

```python
plt.figure(figsize=(10,5))
counts=df["Area"].value_counts()
colors=plt.cm.coolwarm(range(len(counts)))
plt.bar(counts.index,counts.values,color=colors)
plt.title("Countplot of Area")
plt.xlabel("Area")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df["Region"],hue=df["Area"],palette=["black","yellow"])
plt.title("Countplot of Region based on Area")
plt.xticks(rotation=90)
plt.legend(loc="lower right")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(y = df[" Estimated Unemployment Rate (%)"], x = df["Year"], color ="green")
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=df[" Estimated Unemployment Rate (%)"],y=df["Year"],data=df)
plt.title("Unemployment Rate Vs Year")
plt.ylim(2018,2021)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=df[" Estimated Labour Participation Rate (%)"],y=df["Region"],color="green")
plt.title("Labour Participation Rate Vs Region")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=df[" Estimated Unemployment Rate (%)"],y=df["Region"],hue=df["Year"],palette=["blue","pink"])
plt.title("Unemployment Rate Vs Region")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=df["Date"],y=df[" Estimated Labour Participation Rate (%)"],hue=df["Area"],data=df)
plt.title("Percentage of Labour Participation Rate (%)")
plt.ylabel("Labour Participation Rate")
plt.show()
```
